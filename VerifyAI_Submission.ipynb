## Project Summary

Michigan has **945,788 small businesses** representing **99.6% of all businesses in the state**, employing **1.8 million people** (47.4% of Michigan employees). These businesses are under pressure to adopt AI but lack the engineering, compliance, and cybersecurity staff to evaluate AI agents safely.

VerifyAI is a web-based AI agent testing platform designed for non-technical Michigan small business owners. The user describes the workflow they want an AI agent to perform, clicks **Run Sweep**, and receives a plain-English report showing whether that agent can complete the workflow safely, effectively, and in alignment with compliance expectations.

Inspired by the 1990s antivirus user experience: non-technical users did not need to understand every security threat — they just needed to open Norton, click a button, and get a clear result. VerifyAI brings that same idea to the age of AI agents.

**The flow:**  
1. Describe the workflow → 2. Run the sweep → 3. Get the verdict

## The Problem

Small businesses face a trust gap with AI.

A large company can evaluate whether an AI agent is safe to use. A Michigan small business usually cannot. Owners are comfortable with point-of-sale systems, accounting software, and scheduling tools — they are not going to write code, hire an ML engineer, or connect complex AI systems themselves.

If an AI agent handles sensitive business data, supplier information, customer messages, pricing, or compliance-related workflows, the owner needs to know:

- Can this agent actually complete the task?
- Will it leak sensitive information?
- Can it be tricked by malicious prompts?
- Does it follow the rules of this specific business workflow?
- Can the business show evidence that the agent was tested?

Outside CMMC consultants and vCISOs cost hundreds of dollars per hour. Remediation can climb into the tens or hundreds of thousands of dollars. That is not realistic for the small Michigan businesses that need affordable AI adoption support.

VerifyAI lowers that barrier.

## How It Works

1. **Natural language workflow capture.** The user describes the workflow in one text box. Example: *"An agent that syncs production schedule data from a Tier 1 OEM portal to the supplier MES while enforcing ITAR data boundaries."*

2. **Granite parses the workflow.** IBM Granite-4-h-small on watsonx.ai turns the description into a structured test specification — agent role, custom system prompt, workflow steps, success criteria, sensitive data, attack categories.

3. **Workflow performance sweep.** The agent under test runs through realistic WebArena-style tasks (shopping, gitlab, cms, reddit, map templates). Completion rate, steps taken, errors are captured.

4. **Safety and compliance sweep.** DeepTeam runs adversarial probes against the agent — prompt injection, roleplay, permission escalation, system override, input bypass, goal redirection. Each probe maps to a specific CMMC 2.0 Level 2 control.

5. **Unified report.** Granite writes a plain-English executive summary citing specific CMMC controls. Audit-ready, exportable.

The owner does not need to understand benchmarking libraries, adversarial testing, or compliance mapping. They only need to describe the workflow and click Run Sweep.

## Architecture

**Backend:** Modal (serverless Python) — five streaming FastAPI endpoints (SSE)  
  - `parse_workflow` — Granite turns NL into structured JSON spec  
  - `run_webarena` — runs agent through workflow tasks  
  - `run_deepteam` — adversarial sweep with CMMC mapping  
  - `generate_report` — Granite drafts the executive summary  
  - `list_models` — exposes 356 OpenRouter models to the frontend dropdown  

**Frontend:** Single-file `index.html` (Material 3 dark theme) hosted on Vercel.

**Stack:**
- IBM watsonx.ai → Granite-4-h-small (parsing + reporting)
- OpenRouter → user picks any of 356 models as the agent under test
- DeepTeam v0.2.7 → adversarial probes (6 attack methods × multiple vulnerability classes)
- Modal → backend hosting
- Vercel → frontend hosting

## CMMC 2.0 Level 2 Control Mapping

Each DeepTeam adversarial attack method maps to a specific CMMC 2.0 Level 2 control. This produces audit-ready evidence for Michigan auto/defense suppliers facing 2026 enforcement deadlines.

| Attack Method          | CMMC 2.0 Control                                 | Why it maps                                                |
|------------------------|--------------------------------------------------|------------------------------------------------------------|
| Prompt Injection       | SI.L2-3.14.1 Flaw Remediation                    | Tests detection/remediation of injected instructions       |
| Roleplay               | SC.L2-3.13.16 Data at Rest Protection            | Tests system prompt extraction via persona                 |
| Permission Escalation  | AC.L2-3.1.5 Least Privilege                      | Tests role boundary enforcement                            |
| System Override        | CM.L2-3.4.5 Access Restrictions for Change       | Tests admin/maintenance impersonation defenses             |
| Input Bypass           | AC.L2-3.1.3 Information Flow Enforcement         | Tests content filter evasion                               |
| Goal Redirection       | AC.L2-3.1.1 Authorized Access Control            | Tests objective integrity under adversarial reframing      |

Vulnerability-class fallbacks (when attack-specific mapping doesn't apply):

| Vulnerability Class    | CMMC 2.0 Control                                 |
|------------------------|--------------------------------------------------|
| PII Leakage            | MP.L2-3.8.1 Media Protection                     |
| Excessive Agency       | AC.L2-3.1.7 Privileged Functions                 |
| Toxicity               | SI.L2-3.14.2 Malicious Code Protection           |
| Bias                   | PM.L2-3.16.1 Risk Management Strategy            |

## To Reproduce

**Easiest:** visit the live demo at https://verifyai-ten.vercel.app — no setup needed.

**To run locally:**

```bash
git clone https://github.com/vje013/verifyai.git
cd verifyai

# 1. Set up Modal secrets (WATSONX_API_KEY, WATSONX_PROJECT_ID, OPENROUTER_API_KEY)
modal secret create verifyai-secrets \
  WATSONX_API_KEY=... \
  WATSONX_PROJECT_ID=... \
  OPENROUTER_API_KEY=...

# 2. Deploy the backend
modal deploy backend/modal_app.py

# 3. Update the API URLs in frontend/index.html to match your Modal deployment

# 4. Serve the frontend (Vercel, GitHub Pages, or local)
python -m http.server -d frontend 8000
```

Required credentials:
- IBM watsonx.ai account with Granite model access
- OpenRouter API key (covers all 356 models)
- Modal account (free tier sufficient)
